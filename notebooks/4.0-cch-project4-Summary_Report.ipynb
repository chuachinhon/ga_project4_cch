# WHAT THE DATA ON DATA JOBS TELL US
Data science-related jobs are among the most talked about in the market these days, fuelled by both buzz around the industry and actual demand from a growing range of companies. Here's what an analysis of 812 job postings tell us about what employers want, and what job seekers should expect.


## BACKGROUND
The data for this report came via a web-scrape of the Government's jobs bank, [MyCareersFuture](https://www.mycareersfuture.sg/) on April 27, 2019. Some links have expired, as job postings have a limited shelf life. So if you re-run the code on the web-scrape, make sure you don't overwrite the original CSV file.

Elsewhere, I have also removed salary outliers and unsuitable postings to make the dataset more representative. Web-scraping is not a precise activity, and it naturally picked up job postings which had some of the key words we are looking for but are in fact unsuitable, such as auditors and restaurant managers. 

Finally, I also changed the salary data from a number of companies from annual to monthly renumeration.


## SUMMARY
In a nutshell, employers are predominantly looking for experienced data science professionals who also understand the business needs and can lead a team. These senior roles come with higher pay, particularly in the banking and IT industries.

Experienced data professionals - especially senior data scientists - would reap the most benefits from the current job market outlook, while those entering the industry with little or no experience should temper their expectations. Entry level and junior data professionals can expect a monthly pay of under SGD5,000.

The median pay for those taking up senior data roles is above SGD8,000, though there are companies who are willing to pay more for top talent. Let's look at the trends for employers and job seekers in closer detail.


## I. WHAT ARE EMPLOYERS LOOKING FOR:

### 1. Wanted: Data Scientists, Data Engineers
These are the two most commonly seen job titles among the postings (Chart 2). 
While this reflects the nature of the demand by employers, they account for just about 10% of the postings. This is mostly due to the messy and myriad ways in which job titles are used to describe the same role in the industry.




### 2. Experience, business knowledge, management skills
"Data" was the top word used in how companies described the responsibilities of of potential hires (Chart 12). 

But below that, the top key words used were: Business, team work (showed up as two separate words), management, solutions, teams etc. This is consistent with what we know informally from those already working in the industry - that companies don't just want data experts, but also those who can translate their expertise to deal with business needs.



And when it comes to job requirements, employers use the word "experience" most often, followed by expected terms like "skills" and "knowledge" (Chart 13A). 



This could be a potential stumbling block for newcomers, though "experience" can be interpreted somewhat widely in an industry that sees substantial crossovers from those who are not native to the tech or STEM industry.


### 3. Technical skills/background: Python, SQL, Computer Science, Engineering   
Surprisingly, terms used to describe technical skills only start appearing within the top 20 list of key words used in job requirements. "Engineering" is the top technical skills term, followed by "Python", "analytics", "computer science" and "SQL" (Chart 13B).



Terms referencing more advanced skills, such as "machine learning" and "big data", appear further down the list, though still within the top 50 most used words in descriptions of job requirements. "Deep learning" falls outside of the top 50 list, reflecting its relative niche status in the current job market. 

See Chart 14 for a side-by-side comparison of the top words used to describe job responsibilities and desired technical skills in data roles.




## II. WHERE ARE THE JOBS (AND HIGH PAY):

### 1. Median monthly salary: SGD 6,750
Some employers looking to fill senior roles are prepared to pay over SGD20,000 in some instances. But the median monthly salary for data professionals is a more modest SGD6,750 (Chart 1). About a third (35%) of the average salaries offered fall between SGD4,50 and SGD7,500. 




### 2. Banks pay best, most data jobs in IT sector
Data jobs in the banking and finance sectors offer the best median salaries, while most of the openings (40%) are in the IT sector (which also offers competitive salaries, as Chart 9 shows).



Chart 10 below shows DBS Bank as the company offering the highest median salaries for its data roles, followed closely by accounting firm Ernst & Young and Traveloka, an Indonesian travel booking site.



If you are fussy about where you work, well, you'll be happy to know that popular office locations in the CBD and Marina Bay are home to some of the best paying data jobs. Those with the right skills and are prepared to travel a little further out west to Science Park Drive will find two high-paying jobs by Merck and Johnson & Johnson.




## III. WHAT JOB SEEKERS SHOULD KNOW:

### 1. If you are experienced, you are in demand and will command more pay
This is obvious, but it's always good to have visual confirmation and a sense of what the pay gap looks like between experienced and inexperienced hires:




### 2. Good job mobility for those "in the middle"
Data professionals who are in the middle of the pack - some years of experience but not yet considered a veteran - will be cheered by the fact that most openings (almost 74%) are for mid-level positions. This will allow junior data professionals the opportunities to move up, and for those already in the middle category to move laterally across different sectors, or perhaps move up to a senior role.




### 3. Manage your expectations if you are a newcomer
Charts 16 and 17 also show why newcomers need to be realistic about their job hunt and salary expectations. Of the 812 data job openings, only 90, or 11%, are for junior or entry-level positions. The median salary is also the lowest, naturally, under SGD5,000 a month. 


### 4. Internships and temporary work arrangements are rare
Data professionals are expected to work with substantial amounts of sensitive data. This is one possible reason why the vast majority of job openings are for permanent roles, with very few opportunities for internships or temporary jobs. But it could well be that companies don't officially advertise for these roles as much.
