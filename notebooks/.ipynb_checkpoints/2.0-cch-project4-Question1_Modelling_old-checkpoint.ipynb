# QUESTION 1: Predicting high salaried jobs in the data industry
I'll address Question 1 by building models to predict high salaries, which is what job seekers are interested in. Who cares about predicting for low salaries, right?. 

Since there's no standard definition of "high salary" for the data industry, I'll define one: 1 standard deviation above the median for average salary, ie, 9,850 Sing dollars (6,750 (median) + 3099.8 (std)).

Next, I'll create a new col indicating whether a job posting has a high salary, or not.  

```python
jobs['High_Salary_Or_Not'] = np.where(jobs['Salary_Average'] > 9850, 1, 0)
```

```python
jobs['High_Salary_Or_Not'].value_counts()
```

```python
jobs.head()
```

#### NOTE: 
We are naturally dealing with an imbalanced class here, as high-paying jobs are obviously harder to come by - just 25% of the dataset. Subsequent modelling and analysis of the confusion matrices will have to take note of this. 

Let's start with four predictors identified from the EDA phase, and progressively move to more complex models with more features.

```python
predictors = ['Company', 'Title', 'Seniority', 'Category']
```

```python
# Since I'll be using a non-linear model like Random Forest, I won't apply the drop_first parameter 
job_predictors = pd.get_dummies(jobs[predictors])
```

```python
X = job_predictors
y = jobs['High_Salary_Or_Not']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
print((X_train.shape))
print((X_test.shape))
```

```python
print(y_test.value_counts())
# The most important benchmark for the models would be the Recall score, 
# ie, how many of the 41 high salaries in the target test set can it predict?
```

#### 1.1 BUILDING BASELINE PERFORMANCE VIA DUMMY CLASSIFIER

```python
dummy = DummyClassifier(strategy='most_frequent')
```

```python
dummy.fit(X_train, y_train)
```

```python
dummy_pred = dummy.predict(X_test)
```

```python
print("Accuracy Score:", accuracy_score(y_test, dummy_pred))
print("Precision Score : ", precision_score(y_test, dummy_pred))
print("Recall Score:", recall_score(y_test, dummy_pred))
print("f1 Score:", f1_score(y_test, dummy_pred))
```

#### BASELINE INDICATORS:

We can expect subsequent classifier models to score high on the accuracy score, given that the dummy model's accuracy score is nearly 0.8. But in this case, we know that the Recall and f1 scores are more important as we are trying to predict for high salaries, instead of average or low-paying salaries which form the majority class. 

#### 1.2 TRYING A TRIO OF MODELS

```python
models = {
    "Log_Reg": (
        LogisticRegression(),
        {"penalty": ["l1", "l2"], "C": np.logspace(-4, 4, 40), "solver": ["liblinear"]},
    ),
    "LinearSVC": (
        LinearSVC(),
        {"C": [1.0], "class_weight": [None], "loss": ["squared_hinge"]},
    ),
    "Random Forest": (
        RandomForestClassifier(),
        {
            "n_estimators": [200],
            "max_depth": [None],
            "max_features": ["auto"],
            "criterion": ["gini"],
        },
    ),
}
```

```python
for name, (model, parameters) in models.items():
    print("----- {} -----".format(name))
    gs = GridSearchCV(model, parameters, cv=5, verbose=0, n_jobs=-1, scoring="roc_auc")
    gs.fit(X_train, y_train)
    print("Best Parameters:", gs.best_params_)
    print("Best Score:", gs.best_score_)
    y_pred = gs.predict(X_test)
    print("Precision Score :", precision_score(y_test, y_pred))
    print("Recall Score:", recall_score(y_test, y_pred))
    print("f1 Score:", f1_score(y_test, y_pred))
    print("")
```

The accuracy scores from the three models are good, but their recall scores are still too low. I won't proceed further with them and will try a Random Forest Classifier with class weightage instead. This method had proven to be effective in Project 3 in detecting outliers.  

#### 1.3 RFC WITH CLASS WEIGHTAGE

```python
weighted_grid = dict(
    n_estimators=[10, 50, 250, 500],
    criterion=['gini'],
    max_features=['auto'],
    max_depth=[None]
)
```

```python
weighted_rf = RandomForestClassifier(random_state=0, n_jobs=-1, class_weight={0:1,1:75})
```

```python
weighted_clf = GridSearchCV(weighted_rf, weighted_grid, cv=5, scoring='f1', verbose=1, n_jobs=-1)
```

```python
weighted_clf.fit(X_train, y_train)
```

```python
weighted_clf.best_params_
```

```python
weighted_clf.best_score_
```

```python
y_pred_weighted = weighted_clf.predict(X_test)
```

```python
print("Accuracy Score:", accuracy_score(y_test, y_pred_weighted))
print("Precision Score :", precision_score(y_test, y_pred_weighted))
print("Recall Score:", recall_score(y_test, y_pred_weighted))
print("f1 Score:", f1_score(y_test, y_pred_weighted))
```

```python
cm = confusion_matrix(y_test, y_pred_weighted)
cm
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="gist_gray_r")
sns.set(font_scale=1.8)
plt.title("Predictions v Actual ")
plt.ylabel("Actual ")
plt.xlabel("Predicted ")
plt.tight_layout()
```

### ANALYSIS:
This looks like a better outcome, with a good balance between predicting more true positives (33 out of 41 high salaried jobs) while maintaining overall accuracy score at a decent level. While we can push the class weightage further, that would come at a cost of creating more "false alarms" or false positives, ie, predicting a job to be high paying when it actually isn't. This would hurt the model's credibility for users.

#### 1.6 USING THE RESPONSIBILITIES COLUMN
Let's see if using a CountVectorizer-generated feature set for the Responsibilities columns would yield better results.

```python
X_resp = jobs['Responsibilities']
y_resp = jobs['High_Salary_Or_Not']
```

```python
X_resp_train, X_resp_test, y_resp_train, y_resp_test = train_test_split(X_resp, y_resp, random_state=42)
```

```python
vect = CountVectorizer(stop_words='english', max_features=200, ngram_range=(1,2))
```

```python
X_train_dtm = vect.fit_transform(X_resp_train)
X_test_dtm = vect.transform(X_resp_test)
```

```python
models_resp = {
    "Naive Bayes": (MultinomialNB(), {"alpha": [1.0], "fit_prior": [True]}),
    "LogReg": (
        LogisticRegression(),
        {"penalty": ["l1", "l2"], "C": np.logspace(-4, 4, 40), "solver": ["liblinear"]},
    ),
    "LinearSVC": (
        LinearSVC(),
        {"C": [1.0], "class_weight": [None], "loss": ["squared_hinge"]},
    ),
}
```

```python
for name,(model,parameters) in models_resp.items():
    print('----- {} -----'.format(name))
    gs = GridSearchCV(model, parameters, cv=5, 
                      verbose=0, n_jobs=-1, scoring='roc_auc')
    gs.fit(X_train_dtm, y_resp_train)
    print("Best Parameters:", gs.best_params_)
    print("Best Score:", gs.best_score_)
    y_pred = gs.predict(X_test_dtm)
    print("Precision Score :", precision_score(y_resp_test, y_pred))
    print("Recall Score:", recall_score(y_resp_test, y_pred))
    print("f1 Score:", f1_score(y_resp_test, y_pred))    
    print('')
```

#### 1.8 COMBINING COUNTERVECTORIZER FEATURE SET WITH OTHER FEATURES
Finally, let's try combining a countervectorizer set, using the Responsibilities column, with the four initial predictors to see if there's any difference to performance. 

```python
vect = CountVectorizer(stop_words="english", max_features=200, ngram_range=(1,2))
X_train_dtm = vect.fit_transform(X_resp_train)
X_test_dtm = vect.transform(X_resp_test)
```

```python
print((X_train_dtm.shape))
print((X_test_dtm.shape))
```

```python
# converting the earlier X_train, X_test dummified set to sparse matrices so that they 
#can be horizontally stacked with the DTM set above
X_train_extra = sp.sparse.csr_matrix(X_train)
X_test_extra = sp.sparse.csr_matrix(X_test)
```

```python
print((X_train_extra.shape))
print((X_test_extra.shape))
```

```python
# We can't con concat the two sets in this case as they are no longer pandas DFs
X_train_full = hstack([X_train_dtm, X_train_extra]).toarray()
X_test_full = hstack([X_test_dtm, X_test_extra]).toarray()
```

```python
print(X_train_full.shape)
print(X_test_full.shape)
```

```python
models_full = {'Naive Bayes':(MultinomialNB(), {
                    'alpha':[1.0],
                    'fit_prior':[True]
                }),
          'LogReg':(LogisticRegression(), {
                        'penalty':['l1','l2'],
                        'C':np.logspace(-4, 4, 40),
                        'solver':['liblinear']
                    })
         }
```

```python
for name,(model,parameters) in models_full.items():
    print('----- {} -----'.format(name))
    gs = GridSearchCV(model, parameters, cv=5, 
                      verbose=0, n_jobs=-1, scoring='roc_auc')
    gs.fit(X_train_full, y_train)
    print("Best Parameters:", gs.best_params_)
    print("Best Score:", gs.best_score_)
    y_pred = gs.predict(X_test_full)
    print("Precision Score :", precision_score(y_test, y_pred))
    print("Recall Score:", recall_score(y_test, y_pred))
    print("f1 Score:", f1_score(y_test, y_pred))    
    print('')
```

#### The results are disappointing. The recall scores are worse than ever, after combining the two sets of predictors. Let's try one last model, using the weighted RFC model with the combined set of predictors.
