### 2. ANALYSING THE RESPONSIBILITIES AND REQUIREMENTS COLUMNS
These text-heavy columns will require the use of NLP techniques for visusalisation. But the text is relatively straightforward without much by way of linguistic or contextual ambiguities to parse.

#### 2.1 TOP WORDS

```python
jobs.head()
```

```python
vect = CountVectorizer(stop_words='english', min_df=10, ngram_range=(1,2))
vect.fit(jobs['Responsibilities'])

print('# of features: {}'.format( len(vect.get_feature_names()) ))
```

```python
responsibility = vect.get_feature_names()
responsibility
```

```python
resp = pd.DataFrame(vect.transform(jobs['Responsibilities']).todense(),
                       columns=responsibility)
```

```python
responsibility_count = resp.sum(axis=0)
responsibility_count.sort_values(ascending = False).head(20)
```

```python
top_responsibility = responsibility_count.sort_values(ascending = False).head(20)
```

```python
top_responsibility.plot(kind="bar", figsize=(16, 10))
plt.title('Chart 12: Top 20 words used to describe responsibilities in data roles', fontsize=20)
plt.ylabel('Frequency', fontsize=16)
plt.show()
```

```python
cvec = CountVectorizer(stop_words='english', min_df=10, ngram_range=(1,2))
cvec.fit(jobs['Requirements'])
print('# of features: {}'.format( len(cvec.get_feature_names()) ))
```

```python
requirements = cvec.get_feature_names()
requirements
```

```python
req = pd.DataFrame(cvec.transform(jobs['Requirements']).todense(),
                       columns=requirements)
```

```python
requirement_count = req.sum(axis=0)
requirement_count.sort_values(ascending = False).head(20)
```

```python
top_requirement = requirement_count.sort_values(ascending = False).head(20)
```

```python
top_requirement.plot(kind="bar", figsize=(16, 10))
plt.title('Chart 13: Top 20 words used to describe requirements in data roles', fontsize=20)
plt.ylabel('Frequency', fontsize=16)
plt.show()
```

"Experience" is the dominant word in requirements stated by employers, accounting for 35% of the keywords picked up by the countvectorizer. This gives important context to the analysis, and is in line with industry talk about the lack of experienced data professionals (and hence the high demand for them). Let's try to put Charts 12 and 13 side-by-side for a better comparison. 

```python
fig, axs = plt.subplots(1,2, figsize=(16,10), sharey=True)
top_responsibility.plot(kind="bar", figsize=(16, 10), ax=axs[0])
top_requirement.plot(kind="bar", figsize=(16, 10), ax=axs[1])
fig.suptitle(
    "CHART 14: Top 20 words cited in responsibilities (left) and requirements (right) for data roles",
    fontsize=16,
)
plt.show()
```

From the top words used in the responsibilities and requirements columns, it is not hard to interpret what employers are looking for: Experienced and knowledgeable data professionals who know the business needs, and can foster team work.

#### 2.2 TF-IDF SCORES

#### 2.2.1 RESPONSIBILITIES

```python
vect = TfidfVectorizer(stop_words='english', ngram_range=(1,2), sublinear_tf=True)
```

```python
dtm = vect.fit_transform(jobs['Responsibilities'])
features = vect.get_feature_names()
dtm.shape
```

```python
tfidf = pd.DataFrame(dtm.toarray(), columns=features)
tfidf.head()
```

```python
doc = 0
feature_index = dtm[doc,:].nonzero()[1]
tfidf_scores = zip(feature_index, [dtm[doc, x] for x in feature_index])
```

```python
for w, s in [(features[i], s) for (i, s) in sorted(tfidf_scores, key=lambda x: x[1], reverse=True)]:    
    print(w, s)
```

Using a n-grams size of 2, we get a beter sense of the top word-pairings recruiters use in their postings:
- work streams 0.16561206120270844
- problem solver 0.15093472293581758
- technology platforms 0.15093472293581758
- curious mindset 0.10855027832436853
- high execution 0.10855027832436853
- execution bias 0.10855027832436853
- engineering domain 0.10855027832436853
- domain knowhows 0.10855027832436853

This adds to the growing picture we are able to build for what recruiters are looking for. Let's try the TF_IDF scores for Requirements.

### 3. STATS MODEL CHECK-IN BEFORE MODELLING
From the data visualisation, we can see that median average salaries vary signficantly according to company, title, seniority, category and address. The only category showing minnor variation is employment type. Let's test this out using the stats model approach.

```python
f = "Salary_Average ~ C(Company) + C(Title) + C(Seniority) + C(Category) + C(Address)"
model = smf.ols(f, data=jobs)
results = model.fit()
```

```python
results.summary()
# R2 score of 0.977 and adjusted R2 of 0.781
```

The R2 score is suspiciously high, and the gap with the adjusted R2 quite significant. Let's try to remove one predictor - address.

```python
f1 = "Salary_Average ~ C(Company) + C(Title) + C(Seniority) + C(Category)"
model1 = smf.ols(f1, data=jobs)
results1 = model1.fit()
```

```python
results1.summary()
# R2 score of 0.976 and adjusted R2 of 0.785
```

The adjusted R2 score improves with the removal of address. For modelling stage, I'll stick with an initial model of just four predictors then - company, title, seniority and category.
